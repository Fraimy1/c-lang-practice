```python
num = input('enter a number: ')
num = int(num)

def dec_to_bin(n:int, res:int = 0):

    if n<=0:
        return res
    
    power = len(str(res)) - 1
    res += n%2 * 10**power
    return dec_to_bin(n//2, res)
        


dec_to_bin(num, 0)
```
